```python
import pandas as pd

# Assuming 'ratings' is a list, convert it to a DataFrame
ratings = pd.DataFrame(ratings)

# Now you can perform DataFrame operations on 'ratings'
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
rating_with_books=ratings.merge(books,on="ISBN")
```

```python
num_rating=rating_with_books.groupby('Book-Title')['rating'].count().reset_index()
```

```python
num_rating.head()
```

```python
num_rating.rename(columns={'rating':'num_of_rating'},inplace=True)
```

```python
num_rating.head()
```

```python
rating_with_books.head()
```

```python
rating_with_books.merge(num_rating,on='Book-Title')
```

```python
final_ratings= rating_with_books.merge(num_rating,on='Book-Title')
```

```python
final_ratings.head()
```

```python
final_ratings.shape
```

```python
final_ratings[final_ratings['num_of_rating']>=50]
```

```python
final_rating=final_ratings[final_ratings['num_of_rating']>=50]
```

```python
final_rating.head()
```

```python
final_rating.sample(10)
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','Book-Title'],inplace=True)
```

```python
final_rating.shape
```

```python
final_rating
```

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='Book-Title',values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse=csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance,suggestion=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
def recommend_book(book_name):
   book_id=np.where(book_pivot.index==books_name)[0][0]
   distance,suggestion=model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)
   for i in range(len(suggestion)):
        books=book_pivot.index[suggestion[i]]
        for j in books:
            print(j)
```

```python
books_name='A Bend in the Road'
recommend_book(books_name)
```
